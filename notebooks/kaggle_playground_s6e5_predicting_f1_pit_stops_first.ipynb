```python
from catboost import CatBoostClassifier
from sklearn.model_selection import GroupKFold
from sklearn.metrics import roc_auc_score
import numpy as np
```

```python
train = train.dropna(subset=["PitNextLap"])
```

```python
print(train["PitNextLap"].isna().sum())
```

```python
target = "PitNextLap"
features = [col for col in train.columns if col not in ["id",target]]
X = train[features]
y = train[target]
categorical_features = ["Driver", "Compound", "Race"]
print(features)
```

```python
gkf = GroupKFold(n_splits=5)

scores = []

for fold, (train_idx, valid_idx) in enumerate(
    gkf.split(X, y, groups)
):

    print(f"\n========== FOLD {fold + 1} ==========")

    X_train = X.iloc[train_idx]
    X_valid = X.iloc[valid_idx]

    y_train = y.iloc[train_idx]
    y_valid = y.iloc[valid_idx]

    model = CatBoostClassifier(
        iterations=500,
        learning_rate=0.05,
        depth=6,
        eval_metric="AUC",
        verbose=100
    )

    model.fit(
        X_train,
        y_train,
        cat_features=categorical_features,
        eval_set=(X_valid, y_valid),
        use_best_model=True
    )

    preds = model.predict_proba(X_valid)[:, 1]

    auc = roc_auc_score(y_valid, preds)

    print(f"Fold AUC: {auc}")

    scores.append(auc)

print("\n========== RESULTS ==========")
print("Mean AUC:", np.mean(scores))
print("Std AUC:", np.std(scores))
```

```python
feature_importance = model.get_feature_importance()

importance_df = pd.DataFrame({
    "Feature": features,
    "Importance": feature_importance
})

importance_df = importance_df.sort_values(
    by="Importance",
    ascending=False
)

print(importance_df)
```

```python
features_no_year = [
    col for col in features if col != "Year"
]

X_no_year = train[features_no_year]
```

```python
gkf = GroupKFold(n_splits=5)

scores_no_year = []

for fold, (train_idx, valid_idx) in enumerate(
    gkf.split(X_no_year, y, groups)
):

    print(f"\n========== FOLD {fold+1} ==========")

    X_train = X_no_year.iloc[train_idx]
    X_valid = X_no_year.iloc[valid_idx]

    y_train = y.iloc[train_idx]
    y_valid = y.iloc[valid_idx]

    model = CatBoostClassifier(
        iterations=500,
        learning_rate=0.05,
        depth=6,
        eval_metric="AUC",
        verbose=0
    )

    model.fit(
        X_train,
        y_train,
        cat_features=["Driver", "Compound", "Race"],
        eval_set=(X_valid, y_valid),
        use_best_model=True
    )

    preds = model.predict_proba(X_valid)[:, 1]

    auc = roc_auc_score(y_valid, preds)

    print("Fold AUC:", auc)

    scores_no_year.append(auc)

print("\nMean AUC:", np.mean(scores_no_year))
print("Std AUC:", np.std(scores_no_year))
```
